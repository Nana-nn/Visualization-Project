```python
import pandas as pd

from pyecharts import options as opts
from pyecharts.globals import ThemeType
from pyecharts.charts import Timeline, Bar, Map, Pie, Grid
t1 = Timeline(    # 设置画布大小，主题
        init_opts=opts.InitOpts(width="1230px", height="580px", theme=ThemeType.PURPLE_PASSION)
    )
for i in range(2016, 2021):
    data=[[[x["name"], x["value"]] for x in d["data"]] for d in map_data if d["time"] == f'{i}年'][0]
    map0 = (
        Map()
        .add("", data, maptype="美国",label_opts=opts.LabelOpts(
                    is_show=False # 不显示地名
                ),)
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(dimension=0,max_=max([num[1][0] for num in data]),min_=min([num[1][0] for num in data]),range_text=['High','Low'],
                        textstyle_opts=opts.TextStyleOpts(
                        color="rgba(0,0,0,0.5)",  # 文字颜色
                        font_size=15
                    )),
        )
    )
    chart_bar = (
        Bar().add_xaxis(
            # 格式为[1,2,3,4,5]
            [item[0] for item in data]  # 传入x轴的数据
        ).add_yaxis(
            series_name="",  # 系列名称
            y_axis=[item[1][0] for item in data],  # y轴的数据
            label_opts=opts.LabelOpts(
                font_size=9,
                # 标签配置
                position='right',  # 标签显示的位置，inside，ledt，top，bottom
                formatter="{b}: {c}"  # 标签显示的格式
            )).reversal_axis()  # x轴y轴倒置
        .set_global_opts(
            # 不显示坐标轴的 标签
            xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(
                is_show=False)),
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(
                is_show=False)),
            tooltip_opts=opts.TooltipOpts(is_show=False),  # 关闭提示
             title_opts=opts.TitleOpts(
                title='美国各州'+str(i)+'年发生的交通事故数量',
                pos_top='top',
                pos_left='center',
                title_textstyle_opts=opts.TextStyleOpts(
                    font_size=35,
                    color="rgba(0,155,155,0.9)"
                ),
            ),
            visualmap_opts=opts.VisualMapOpts(
                is_calculable=True,  # 显示组件拖拽手柄
                dimension=0,
                pos_left='0%',  # 组件位置
                pos_bottom='66%',  # 组件位置
                range_text=['max', 'min'],  # 组件两端的文本
                range_color=["lightskyblue", "yellow", "orangered"],  # 组件过渡颜色
                max_=max([num[1][0] for num in data]),
                min_=min([num[1][0] for num in data]),
                textstyle_opts=opts.TextStyleOpts(
                    color="rgba(0,0。5,0,0.5)",  # 文字颜色
                    #font_size=10
                ),
            ),
        ))
    
    pie_data = [[x[0], x[1][0]] for x in data]
    pie_chart = (
        Pie()
            .add(
                series_name="",
                data_pair=pie_data,
                radius=["10%", "20%"],  # 半径大小，相对整个画布高宽的一半，数组的第一项是内半径，第二项是外半径
                center=["90%", "65%"],  # 位置，相对整个画布来说
                itemstyle_opts=opts.ItemStyleOpts(
                    border_width=1, border_color="rgba(0,0,0,0.3)"
                ),
            )
            .set_global_opts(
                tooltip_opts=opts.TooltipOpts(is_show=True, formatter="{b} {d}%"),  # 显示的格式
                legend_opts=opts.LegendOpts(is_show=False),
            )
    )
    grid_chart = (
        Grid()
            .add(chart_bar, grid_opts=opts.GridOpts(pos_left="0", pos_right="40%", pos_top="34%", pos_bottom="40"))
            .add(pie_chart, grid_opts=opts.GridOpts())
            .add(map0, grid_opts=opts.GridOpts())
    )
    t1.add(grid_chart, "{} 年".format(i))
t1.render("grid.html")




```
